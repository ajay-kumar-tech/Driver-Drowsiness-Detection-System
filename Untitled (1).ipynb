```python
import numpy as np
import glob
from tqdm import tqdm
from PIL import Image
import random

X = []
Y = []

open_eyes_path = r"C:\Users\apmot\mrl_dataset\train\Open_Eyes\*.png"
closed_eyes_path = r"C:\Users\apmot\mrl_dataset\train\Closed_Eyes\*.png"

for img_path in tqdm(glob.glob(open_eyes_path)):
    try:
        img = Image.open(img_path).convert("RGB").resize((224, 224))
        X.append(np.array(img))
        Y.append(1)
    except Exception as e:
        print(f"Error loading {img_path}: {e}")

for img_path in tqdm(glob.glob(closed_eyes_path)):
    try:
        img = Image.open(img_path).convert("RGB").resize((224, 224))
        X.append(np.array(img))
        Y.append(0)
    except Exception as e:
        print(f"Error loading {img_path}: {e}")

combined = list(zip(X, Y))
random.shuffle(combined)
X, Y = zip(*combined)
X = np.array(X)
Y = np.array(Y)
```

```python

print("X shape:", X.shape)  # should be (2000, 224, 224, 3)
print("Y shape:", Y.shape)  # should be (2000,)
```

```python
X = X/ 255.0
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
#Deep learning model for training -Transferlearning 
```

```python
from tensorflow.keras.applications import MobileNet
from tensorflow.keras import layers, models

model = MobileNet(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

base_input = model.input
base_output = model.layers[-4].output  # 4th layer from the end

flat = layers.Flatten()(base_output)
dense = layers.Dense(1, activation='sigmoid')(flat)

new_model = models.Model(inputs=base_input, outputs=dense)
```

```python
# #freeze the layeers 
# for layers in new_model.layers:
#     layers.trainable=False
```

```python
new_model.summary()
```

```python
new_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
new_model.fit(X_train, y_train, epochs=2, batch_size=32, validation_split=0.2)
```

```python
new_model.evaluate(X_test, y_test)
```

```python
new_model.save("new_drowsiness_model.h5")
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load Haar cascades
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load the trained drowsiness model
new_model = tf.keras.models.load_model(r"C:\Users\apmot\Downloads\new_drowsiness_model.h5")


# Start video capture
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect eyes
    eyes = eye_cascade.detectMultiScale(gray, 1.1, 4)
    status = "Eyes Not Detected"

    for (x, y, w, h) in eyes:
        roi_color = frame[y:y+h, x:x+w]

        try:
            # Preprocess eye image
            eyes_roi = cv2.resize(roi_color, (224, 224))
            eyes_roi = np.expand_dims(eyes_roi, axis=0)  # Add batch dimension
            eyes_roi = eyes_roi / 255.0

            # Predict
            prediction = new_model.predict(eyes_roi)

            # Interpret prediction
            if prediction[0][0] == 1.0:
                status = "Open Eyes"
            else:
                status = "Closed Eyes"

            # Draw rectangle on detected eyes
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        except Exception as e:
            print(f"Prediction error: {e}")
            status = "Error in Prediction"

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display status text
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, status, (50, 50), font, 3, (0, 255, 0), 2, cv2.LINE_4)

    cv2.imshow('Drowsiness Detection Tutorial', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import winsound
import numpy as np
import cv2
from keras.models import load_model  # or tensorflow.keras.models if needed

# Load your trained model
new_model = tf.keras.models.load_model(r"C:\Users\apmot\Downloads\new_drowsiness_model.h5")

# Sound parameters
frequency = 2500  # Hz
duration = 1000   # ms

# Load Haar cascades
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

# Open webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

counter = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 4)
        for (ex, ey, ew, eh) in eyes:
            eyes_roi = roi_color[ey:ey+eh, ex:ex+ew]
            final_image = cv2.resize(eyes_roi, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)
            final_image = final_image / 255.0

            prediction = new_model.predict(final_image)

            font = cv2.FONT_HERSHEY_SIMPLEX
            if prediction[0][0] == 1.0:
                status = "Open Eyes"
                color = (0, 255, 0)
                counter = 0
            else:
                status = "Closed Eyes"
                color = (0, 0, 255)
                counter += 1

            # Display status
            cv2.putText(frame, status, (150, 150), font, 1.2, color, 2, cv2.LINE_4)

            # If eyes are closed for more than 5 frames, trigger alert
            if counter > 5:
                cv2.rectangle(frame, (0, 0), (300, 75), (0, 0, 0), -1)
                cv2.putText(frame, 'Sleep Alert !!', (10, 50), font, 1.2, (0, 0, 255), 2)
                winsound.Beep(frequency, duration)
                counter = 0
            break  # Only use first detected eye for simplicity

    cv2.imshow('Drowsiness Detection', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import serial
import time
import tensorflow as tf
from tensorflow.keras.models import load_model

# Load the trained model
new_model = tf.keras.models.load_model(r"C:\Users\apmot\Downloads\new_drowsiness_model.h5")

# Initialize Serial Communication with Arduino
arduino = serial.Serial(port='COM8', baudrate=9600, timeout=1)  # Change COM4 if needed
time.sleep(2)  # Wait for Arduino to reset

# Load Haar cascades
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

# Open webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

counter = 0  # Counter to check consecutive closed-eye frames

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 4)

        for (ex, ey, ew, eh) in eyes:
            eye_roi_color = roi_color[ey:ey+eh, ex:ex+ew]

            # Preprocess for model
            final_image = cv2.resize(eye_roi_color, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)
            final_image = final_image / 255.0

            prediction = new_model.predict(final_image, verbose=0)

            font = cv2.FONT_HERSHEY_SIMPLEX
            if prediction[0][0] == 1.0:
                status = "Open Eyes"
                color = (0, 255, 0)
                counter = 0
                arduino.write(b'0')  # Send 0 to Arduino -> Buzzer OFF
            else:
                status = "Closed Eyes"
                color = (0, 0, 255)
                counter += 1

            # Display status
            cv2.putText(frame, status, (10, 40), font, 1.0, color, 2, cv2.LINE_AA)

            # If eyes closed for more than 5 frames, trigger buzzer
            if counter > 5:
                cv2.rectangle(frame, (0, 0), (300, 75), (0, 0, 0), -1)
                cv2.putText(frame, 'Sleep Alert !!', (10, 50), font, 1.2, (0, 0, 255), 2, cv2.LINE_AA)
                arduino.write(b'1')  # Send 1 to Arduino -> Buzzer ON
                counter = 0  # Reset counter after alert

            break  # Only process the first detected eye

    cv2.imshow('Drowsiness Detection', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
arduino.close()
cv2.destroyAllWindows()
```
